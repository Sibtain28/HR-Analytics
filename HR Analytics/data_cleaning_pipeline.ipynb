# HR Data Cleaning Pipeline
This notebook provides a structured approach to cleaning the HR dataset.

## Objectives:
1. Load the raw Excel data.
2. Perform data cleaning (handling duplicates, missing values, etc.).
3. Save the cleaned data for further analysis.

```python
import pandas as pd
import os

# Define paths
raw_data_path = 'dataset/raw/HR Data.xlsx'
clean_data_path = 'dataset/clean/cleaned_dataset.csv'

# Load the dataset
print(f"Loading data from {raw_data_path}...")
df = pd.read_excel(raw_data_path)

# Display first few rows
df.head()
```

```python
# Basic Data Cleaning

print("Starting data cleaning...")

# 1. Remove duplicates
initial_count = len(df)
df = df.drop_duplicates()
print(f"Removed {initial_count - len(df)} duplicate rows.")

# 2. Handle missing values
# Dropping rows where essential columns are null (adjust as needed)
df = df.dropna(how='all')

# 3. Standardize column names (lowercase and underscore)
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

print("Data cleaning completed.")
```

```python
# Save cleaned dataset
if not os.path.exists('dataset/clean'):
    os.makedirs('dataset/clean')

df.to_csv(clean_data_path, index=False)
print(f"Cleaned dataset saved to: {clean_data_path}")
```
